```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_ecom = pd.read_csv("/kaggle/input/ecommerce-data/data.csv", encoding='latin1')
df_ecom.head()
```

```python
df_ecom.tail()
df_ecom.info()
df_ecom.describe()
```

```python
print(df_ecom.isnull().sum())
```

```python
product_sales = df_ecom.groupby("Description")["Quantity"].sum().head(10)

plt.figure(figsize=(8,4))
product_sales.plot(kind='bar')
plt.title("Sales per Product")
plt.show()

product_sales.plot(kind='line')
plt.title("Sales Trend")
plt.show()
```

```python

df_house = pd.read_csv("/kaggle/input/housing-price-prediction/Housing.csv")
df_house.head()
```

```python
print(df_house.isnull().sum())
```

```python
plt.scatter(df_house["area"], df_house["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
numeric_df = df_house.select_dtypes(include=['int64','float64'])

corr = numeric_df.corr()

plt.imshow(corr)
plt.colorbar()
plt.title("Housing Correlation Heatmap")
plt.show()
```

```python

df_bank = pd.read_csv("/kaggle/input/customer-personality-analysis/marketing_campaign.csv", sep='\t')

df_bank.head()
```

```python
print(df_bank.isnull().sum())
df_bank.info()
```

```python
df_bank['Total_Spending'] = (
    df_bank['MntWines'] +
    df_bank['MntFruits'] +
    df_bank['MntMeatProducts'] +
    df_bank['MntFishProducts'] +
    df_bank['MntSweetProducts'] +
    df_bank['MntGoldProds'] )
```

```python
avg_spending = df_bank.groupby('Education')['Total_Spending'].mean()

plt.figure(figsize=(10,6))
plt.bar(avg_spending.index, avg_spending.values)
plt.title("Average Spending by Education Level")
plt.xlabel("Education")
plt.ylabel("Average Spending")
plt.xticks(rotation=45)
plt.show()

plt.boxplot(df_bank["Year_Birth"])
plt.title("Customer Age Distribution")
plt.show()
```
